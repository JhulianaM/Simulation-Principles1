```python
import numpy as np
import matplotlib.pyplot as plt


N=1000
#x=np.random.uniform(0,2,N)
x=np.random.uniform(0.156517,1.156517,N)
e=np.exp(2)-1
Xexp=np.log(x*e)
#print(Xexp)

X=np.arange(0, 2, 0.1)
fx=np.exp(X)/(np.exp(2)-1)
plt.plot(X,fx)
plt.hist(Xexp,bins=30,density=True)
plt.show()
```
