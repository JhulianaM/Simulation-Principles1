```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#p probabilidad de suceso
#x numero de tentativas
p = 20 / (30 + 20)
x=1
x =np.arange(0, 16, 1)
fx=p * (1 - p)**(x - 1)

plt.plot(x, fx, 'b-', lw=2)
plt.title('Distribuição Geometrica')
plt.ylabel('D. Probabilidade')
plt.grid(True)
plt.show()
```

```python
#Generador de datos Distribusion Poisson
data = np.random.geometric(p=20/50, size=10000)
#Histograma
plt.hist(data, bins=20, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de una variable con distribución de Geometrica')
plt.grid(True)
plt.show()
```

```python
# Definindo a probabilidade de sucesso
p = 20 / (30 + 20)  # 20 bolas pretas em um total de 50 bolas

# Gerando uma variável aleatória geométrica
def gerar_variavel_geometrica(p):
    return np.random.geometric(p)

# Calculando a probabilidade de que a 6ª bola seja a primeira preta
probabilidade = p * (1 - p)**(6 - 1)

print(f"A probabilidade de que a 6ª bola retirada seja a primeira preta é: {probabilidade:.4f}")
```
